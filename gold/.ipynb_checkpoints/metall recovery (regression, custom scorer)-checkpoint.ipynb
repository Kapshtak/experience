# Восстановление золота из руды

Основной задачей в рамках данного проекта является построение моделей, предсказывающих коэффициент восстановления золота из золотосодержащей руды на различных этапах производственного процесса: для чернового концентрата и для финального концентрата. 

В вашем распоряжении данные с параметрами добычи и очистки, разбитые на обучающую и тестовую выборки. Также был предоставлен исходный набор данных. Некоторые параметры для обучения моделей нам недоступны, потому что рассчитываются на более поздних этапах производственного процесса. Из-за этого в тестовой выборке отсутствуют некоторые признаки, которые могут быть в обучающей. Также в тестовом наборе нет целевых признаков.

В рамках решения поставленной задачи необходимо было решить ряд дополнительных задач:
- проверить корректность расчета эффективности восстановления руды (наш целевой признак) - необходимо было сделать самостоятельный расчет по приведенной формуле, а также вычислить метрику MAE между данными из датасета, а также результатами нашего расчета;
-  изучить изменение концентрации металлов на разных этапах очистки, исследовать суммарную концентрацию веществ на разных этапах производственного процесса, а также сравнить распределение некоторых из признаков в обучающей и тестовой выборках;
- определить функцию, которая будет использоваться в качестве метрики для оценки качества моделей.

Более подробно посмотрим на пропуски, для чего будем использовать функцию.

```python
def missing_values_table(df):
        # Количество пропущенных значений
        mis_val = df.isna().sum()
        
        # Процентное соотношение пропущенных значений
        mis_val_percent = 100 * df.isna().sum() / len(df)
        
        # Создание датафрейма из полученных данных
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Переименование столбцов
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Отсутствующие значения', 1 : '% от общего количества значений'})
        
        # Фильтрация и сортировка полученного датафрейма
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% от общего количества значений', ascending=False).round(1)
        
        return mis_val_table_ren_columns
```

```python
dates_with_na = []
for i in columns_with_na:
    dates_with_na += list(data_full[data_full[i].isna()]['date_str'].unique())
```

Создадим словарь, ключем которого будет строка с датой, в которую имел место пропуск в каком-либо из интересующих нас значений, а значением - вложенный словарь, в котором ключами будут названия четырех интересующих нас столбцов, в которых мы заполняем пропуски, а значениями - медианное значение интересующего нас показателя за соответствующий день.

```python
for_missing_values = {}
for i in dates_with_na:
        _slice = data_full[data_full['date_str'] == i]
        for_missing_values[i] = {
       'rougher.state.floatbank10_e_air': _slice['rougher.state.floatbank10_e_air'].median(),
       'rougher.input.floatbank11_xanthate': _slice['rougher.input.floatbank11_xanthate'].median(),
       'secondary_cleaner.state.floatbank2_a_air': _slice['secondary_cleaner.state.floatbank2_a_air'].median(),
       'rougher.input.feed_size' : _slice['rougher.input.feed_size'].median()}
```

Создадим функцию, которая будет принимать в качестве аргумента набор данных и наименование столбца, в котором мы заполняем пропуски. Применяться функция будет к строке набора данных. Если дата в строковом формате, указанная в столбце `date_str`, отсутствует в спиcке `dates_with_na`, то функция возвращает текущее значение из столбца, пропуски в котором мы пытаемся заполнить. В противном случае функция будет возвращать медианное значение искомого показателя за соответствующий день.   

```python
def na_filler(data_full, column):
    if data_full['date_str'] not in dates_with_na:
        return data_full[column]
    else:
        return(for_missing_values[data_full['date_str']][column])
```

Применим функцию в цикле, предварительно запомниив количество пропусков до её использования, чтобы оценить результат её работы.

```python
before = missing_values_table(data_full[test_columns])
before
```

```python
for i in columns_with_na:
    data_full[i] = data_full.apply(na_filler, axis=1, column=i)
```

```python
after = missing_values_table(data_full[test_columns])
after
```

```python
print('Изначально было {} пропусков, после работы функции их стало {}, мы заполнили {} ({:.2%}) пропусков'
      .format(before['Отсутствующие значения'].sum(), after['Отсутствующие значения'].sum(),
       before['Отсутствующие значения'].sum() - after['Отсутствующие значения'].sum(),
       (before['Отсутствующие значения'].sum() - after['Отсутствующие значения'].sum()) / 
        before ['Отсутствующие значения'].sum()))
```

Мы избавились от трети пропусков, что уже неплохо. Остальные пропуски заполним с использованием атрибута ffill и limit=12 (половина суток), а оставшиеся пропуски заполним медианным значением по всему набору данных.

```python
for i in columns_with_na:
    data_full[i].fillna(method='ffill', limit=12, inplace=True)
missing_values_table(data_full[test_columns])
```

```python
for i in columns_with_na:
    data_full[i].fillna(data_full[i].median(), inplace=True)
missing_values_table(data_full[test_columns])
```

### Проверка корректности рассчета целевого признака

Далее проверим корректность расчета целевого признака `rougher.output.recovery`, для чего создадим функцию, которая будет принимать в качестве аргумента набор данных, после чего для каждой строки будет рассчитывать эффективность обогащения по заданной формуле. Чтобы избежать исключений, мы добавим конструкцию try/except, во второй части которой мы будем дополнять список `indexes` датой в строковом формате для строки, в которой не удалось произвести расчет эффективности.

```python
indexes = []
def recovery(data_full):
    F = data_full['rougher.input.feed_au' ]
    C = data_full['rougher.output.concentrate_au'] 
    T = data_full['rougher.output.tail_au'] 
    try:
        return (C * (F - T)) / (F * (C - T)) * 100
    except:
        indexes.append(data_full['date_str'])
        pass
```

```python
data_full['recovery_raw_my'] = data_full.apply(recovery, axis=1)
recovery_test = data_full[['date'] + ['recovery_raw_my'] + ['rougher.output.recovery']]
recovery_test.head()
```

C учетом проведенной нами работы по удалению выбросов из столбцов с целевыми признаками, мы добавили в набор данных `data_full` столбец с расчетом эффективности восстановления. Проблем с расчетом не возникло (except в функции ни разу не сработал, поэтому список `indexes` пуст). Посмотрим на среднее абсолютное отклонение для указанного целевого признака.

### Создание кастомной метрики для модели

Напишем функцию для рассчета симметричного среднего абсолютного процентного отклонения, а также с помощью метода `make_scorer()` создадим оценщика для наших моделей. Так как нас интересуют модели с минимальным показателем smape, мы будем использовать атрибут `greater_is_better=False`.

```python
def smape(targets, predictions):
    return abs(1 / len(targets) * np.sum(abs(targets - predictions) / 
                                       ((abs(targets) + abs(predictions)) / 2)) * 100)

scorer = make_scorer(smape, greater_is_better=False)
```

Сразу же напишем функцию, которая позволит осуществлять рассчет итогового sMAPE, чтобы комплексно посмотреть на метрики полученных моделей.

```python
def final_smape(rougher_test_features, rougher_test_targets, rougher_model, 
                final_test_features, final_test_targets, final_model):
    rougher_score = smape(rougher_model.predict(rougher_test_features), rougher_test_targets)
    final_score = smape(final_model.predict(final_test_features), final_test_targets)
    return 0.25 * rougher_score + 0.75 * final_score
```

### Подбор моделей и их гиперпараметров

Определим функцию, которая будет осуществлять перебор нескольких моделей в цикле, и, в зависимости от вида модели, будет осуществлять GridSearch в соответствии переданной для каждого доступного вида модели сеткой параметров (предусмотрено значение по умолчанию). Оценка модели будет осуществляться с помощью созданного выше оценщика, основанного на функции smape. Название лучшей модели, её оценка и гиперпараметры будут возвращаться функцией. В случае, если модель не сможет показать лучший результат, чем Dummy Regressor, будет возвращено соответствующее сообщение.

```python
def modeller(model, features, target, baseline, 
             lr_param=[{'normalize' : [True, False]}],
             rfr_param=[{'n_estimators': range(1, 51, 10), 
                         'max_depth': range(1, 7, 1), 
                         'min_samples_leaf': range(1, 4), 
                         'min_samples_split': range(2, 4, 2)}],
             knn_param=[{'n_neighbors' : range(3, 140, 2),
                         'weights' : ['distance', 'uniform']}],
             metrics='neg_mean_absolute_error'):
    
    best_score = float('inf')
    best_model = None
    best_params = None
    
    for i in model:
        if type(i) == LinearRegression:
            param_grid_lr=lr_param
            grid = GridSearchCV(i, param_grid=param_grid_lr, scoring=metrics, verbose=4)
            grid.fit(features, target)
            score = grid.score(features, target)
            if abs(score) < best_score:
                best_score = abs(score)
                best_model = i
                best_params = grid.best_params_
        
        elif type(i) == RandomForestRegressor:
            param_grid_rfr=rfr_param
            grid = GridSearchCV(i, param_grid=param_grid_rfr, scoring=metrics, verbose=4)
            grid.fit(features, target)
            score = grid.score(features, target)
            if abs(score) < best_score:
                best_score = abs(score)
                best_model = i
                best_params = grid.best_params_
        
        elif type(i) == KNeighborsRegressor:
            param_grid_knn=knn_param
            grid = GridSearchCV(i, param_grid=param_grid_knn, scoring=metrics, verbose=4)
            grid.fit(features, target)
            score = grid.score(features, target)
            if abs(score) < best_score:
                best_score = abs(score)
                best_model = i
                best_params = grid.best_params_
    
    if best_score > baseline:
        print('Ни одна из моделей не показала лучший результат, чем DummyRegressor')
    
    else:
        return best_score, best_model, best_params
```

Создадим экземпляры классов `LinearRegression()` и `RandomForestRegressor()`, для последнего передадим гиперпараметр `random_state=12345`. Поместим созданные экземпляры в список, чтобы передать его в качестве атрибута в функцию `modeller()`.

```python
lr = LinearRegression()
rfr = RandomForestRegressor(random_state=12345)
models = [lr, rfr]
```

Мы выбрали две модели, для которых осуществили подбор параметров. Лучшую оценку показала модель RandomForestRegressor с указанными выше гиперпараметрами. Значение smape для нее на обучающей выборке составило 2.86%. Создадим и обучим модель с найденными гиперпараметрами, чтобы можно было протестировать её на тестовой выборке.
